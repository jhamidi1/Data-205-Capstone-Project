This notebook explores relationships between education, income, and poverty across neighborhoods.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bachelors_col = [col for col in df.columns if "Bachelor" in col][0]
income_col = [col for col in df.columns if "wage or salary income" in col][0]
poverty_col = [col for col in df.columns if "below poverty level" in col][0]
```

```python
bachelors = df[bachelors_col]
income = df[income_col]
poverty = df[poverty_col]
```

```python
[col for col in df.columns if "below poverty level" in col]
```

```python
poverty = df["Estimate; Income in the past 12 months below poverty level:"]
```

## Graph 1

```python
plt.figure(figsize=(8,6))

plt.scatter(bachelors, income)

plt.xlabel("Bachelor's Degree Attainment")
plt.ylabel("Income")
plt.title("Relationship Between Education and Income")

plt.grid(True)
plt.show()
```

Interpretation: Education vs Income

This graph shows the relationship between individuals with bachelor's degrees and the number of people with income across neighborhoods. There appears to be a positive relationship between education and income. Neighborhoods with higher numbers of individuals who have earned a bachelor's degree also tend to have higher numbers of individuals earning income. This shows that education may play an important role in economic outcomes in communities. As education levels increase, more individuals are likely to earn income. However, it is important to keep in mind that this data represents counts of individuals rather than actual income amounts.

```python
plt.figure(figsize=(8,6))

plt.scatter(bachelors, poverty)

plt.xlabel("Bachelor's Degree Attainment")
plt.ylabel("Number of Individuals Below Poverty Level")
plt.title("Education vs Poverty")

plt.grid(True)
plt.show()
```

Interpretation: Education vs Poverty

This graph shows the relationship between the number of individuals with bachelor’s degrees and the number of individuals living below the poverty level across different neighborhoods.
