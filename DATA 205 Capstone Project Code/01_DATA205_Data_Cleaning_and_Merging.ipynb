# Data Ingestion

This notebook loads, cleans, and merges neighborhood level datasets related to education, income, and poverty.

```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
education = pd.read_csv("01_Educational_Attainment.csv")

income = pd.read_csv("02_Aggregate_Household_Income.csv")

poverty = pd.read_csv("03_Poverty_Status.csv")
```

```python
education.head()
income.head()
poverty.head()
```

```python
education.columns = education.columns.str.strip()
income.columns = income.columns.str.strip()
poverty.columns = poverty.columns.str.strip()
```

```python
print(education.columns)
print(income.columns)
print(poverty.columns)
```

```python
merged = education.merge(income, on="Neighborhood", how="inner")
merged = merged.merge(poverty, on="Neighborhood", how="inner")
```

```python
# Create a cleaner final dataset with only the variables needed for analysis

final_data = merged[[
    "Neighborhood",
    "Estimate; Total:_x",
    "Estimate; Male: - Bachelor's degree",
    "Estimate; Female: - Bachelor's degree",
    "Estimate; Total: - With wage or salary income",
    "Estimate; Total:_y",
    "Estimate; Income in the past 12 months below poverty level:"
]].copy()

# Rename the columns to simpler names
final_data = final_data.rename(columns={
    "Estimate; Total:_x": "Adult_Population_25_Over",
    "Estimate; Male: - Bachelor's degree": "Male_Bachelors_Degree",
    "Estimate; Female: - Bachelor's degree": "Female_Bachelors_Degree",
    "Estimate; Total: - With wage or salary income": "Households_With_Wage_Income",
    "Estimate; Total:_y": "Total_Households",
    "Estimate; Income in the past 12 months below poverty level:": "Population_Below_Poverty"
})

# Create the calculated variables
final_data["Total_Bachelors_Degree"] = (
    final_data["Male_Bachelors_Degree"] + final_data["Female_Bachelors_Degree"]
)

final_data["Bachelors_Degree_Pct"] = (
    final_data["Total_Bachelors_Degree"] / final_data["Adult_Population_25_Over"] * 100
)

final_data["Wage_Income_Pct"] = (
    final_data["Households_With_Wage_Income"] / final_data["Total_Households"] * 100
)

final_data["Poverty_Rate"] = (
    final_data["Population_Below_Poverty"] /
    (final_data["Population_Below_Poverty"] + merged["Estimate; Income in the past 12 months at or above poverty level:"])
    * 100
)

# Keep final columns
final_data = final_data[[
    "Neighborhood",
    "Adult_Population_25_Over",
    "Total_Bachelors_Degree",
    "Bachelors_Degree_Pct",
    "Total_Households",
    "Households_With_Wage_Income",
    "Wage_Income_Pct",
    "Population_Below_Poverty",
    "Poverty_Rate"
]]

# Round percentage columns
final_data["Bachelors_Degree_Pct"] = final_data["Bachelors_Degree_Pct"].round(2)
final_data["Wage_Income_Pct"] = final_data["Wage_Income_Pct"].round(2)
final_data["Poverty_Rate"] = final_data["Poverty_Rate"].round(2)

# Preview final cleaned data
final_data.head()
```

```python
final_data.to_csv("04_Final_Merged_Pittsburgh_Neighborhood_Data.csv", index=False)

from google.colab import files
files.download("04_Final_Merged_Pittsburgh_Neighborhood_Data.csv")
```
