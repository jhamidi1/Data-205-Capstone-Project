# Data Ingestion

This notebook loads, cleans, and merges neighborhood level datasets related to education, income, and poverty.

```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
education = pd.read_csv("sex-by-educational-attainment-for-the-population-25-years-and-over (2) (1).csv")

income = pd.read_csv("wage-or-salary-income-in-the-past-12-months-for-households (1).csv")

poverty = pd.read_csv("poverty-status-of-individuals-in-the-past-12-months-by-living-arrangement (1).csv")
```

```python
education.head()
income.head()
poverty.head()
```

```python
education.columns = education.columns.str.strip()
income.columns = income.columns.str.strip()
poverty.columns = poverty.columns.str.strip()
```

```python
print(education.columns)
print(income.columns)
print(poverty.columns)
```

```python
merged = education.merge(income, on="Neighborhood", how="inner")
merged = merged.merge(poverty, on="Neighborhood", how="inner")
```

```python
merged.head()
merged.shape
```

```python
merged.to_csv("merged_data.csv", index=False)
```
